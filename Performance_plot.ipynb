```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data provided
models = ['No Regularizer', 'Linear Entropy Annealing', 'Constant Attention Matrix', 'Factor Analysis']
train_mse = [0.04444, 0.05248, 0.0475, 0.048127]
val_mse = [0.04942, 0.05548, 0.05, 0.046869]
ari = [0.19, 0.23, 1.0, 0.5265]

# Prepare the data in a DataFrame
data = {
    'Model': models * 2,
    'MSE Type': ['Train'] * len(models) + ['Validation'] * len(models),
    'MSE': train_mse + val_mse,
    'ARI': ari + ari  # ARI is the same for train and validation since it's computed per model
}

df = pd.DataFrame(data)

# Set the style
sns.set(style="whitegrid")

# Create the plot
plt.figure(figsize=(12, 8))
palette = sns.color_palette("husl", len(models))

# Plot using Seaborn
sns.scatterplot(
    data=df,
    x='MSE',
    y='ARI',
    hue='Model',
    style='MSE Type',
    s=200,
    palette=palette,
    edgecolor='black'
)

# Customize the plot
plt.title('MSE vs ARI for Different Models', fontsize=16)
plt.xlabel('Mean Squared Error (MSE)', fontsize=14)
plt.ylabel('Adjusted Rand Index (ARI)', fontsize=14)
plt.legend(title='Model and MSE Type', fontsize=12, title_fontsize=13, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```
